```python
import os
import random
import shutil
import re
```

```python
random.seed(20241104)
```

```python

def group_files(files):
    groups = {}
    for file in files:
        # 파일명에서 C1~C8을 제외한 부분을 키로 사용
        origin_name = re.sub(r'_C[1-8]\.mp4$', '', file)
        if origin_name not in groups:
            groups[origin_name] = []
        groups[origin_name].append(file)
    return groups

def copy_random_file_groups(src_dir, dst_dir, fraction=0.1):
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
    
    all_files = [f for f in os.listdir(src_dir) if os.path.isfile(os.path.join(src_dir, f)) and f.endswith('.mp4')]
    file_groups = group_files(all_files)
    
    num_groups_to_select = int(len(file_groups) * fraction)
    selected_groups = random.sample(list(file_groups.keys()), num_groups_to_select)
    
    copied_files = 0
    for group in selected_groups:
        for file in file_groups[group]:
            src_path = os.path.join(src_dir, file)
            dst_path = os.path.join(dst_dir, file)
            shutil.copy(src_path, dst_path)
            copied_files += 1
    
    print(f"{copied_files} files ({len(selected_groups)} groups) copied from {src_dir} to {dst_dir}")

main_dir = 'D:\\human_fall\\video'  
output_dir = 'D:\\human_fall\\re_video'  

for dataset in ["training", "validation"]:
    for class_name in ["N", "Y"]:
        src_dir = os.path.join(main_dir, dataset, class_name)
        dst_dir = os.path.join(output_dir, dataset, class_name)
        
        if os.path.exists(src_dir):
            copy_random_file_groups(src_dir, dst_dir)

print("복사 완료")
```

### 원본 json 파일의 '낙상 시작'과 '낙상 종료' 프레임을 가져오기 위해 기존 json 파일과 같은 이름의 원본 sensor json 파일 가져오기 

```python
# 경로 설정
original_folder = 'D:\\041.낙상사고 위험동작 영상-센서 쌍 데이터\\3.개방데이터\\1.데이터\\Training\\02.라벨링데이터\\TL\\센서\\N' 
processed_folder = 'D:\\human_fall\\re_landmark\\re_train_NY_json' 
new_folder = 'D:\\human_fall\\re_landmark\\sensor_json'

# 새 폴더가 없으면 생성
if not os.path.exists(new_folder):
    os.makedirs(new_folder)

# B 폴더의 파일 목록 가져오기
processed_files = [f for f in os.listdir(processed_folder) if f.endswith('.json')]

# 원본 폴더의 모든 하위 폴더를 순회하며 일치하는 파일 찾아 복사
for root, dirs, files in os.walk(original_folder):
    for file in files:
        if file in processed_files:
            original_file_path = os.path.join(root, file)
            new_file_path = os.path.join(new_folder, file)
            shutil.copy2(original_file_path, new_file_path)
            print("복사중")
            
print("완료")
```
